```python
valueCounts = []
for col in df.columns:
    column_value = df[col].nunique()
    valueCounts.append(column_value)

unique_values_df = pd.DataFrame(valueCounts, index=df.columns, columns=['Total Unique Values'])
print(unique_values_df)
```
