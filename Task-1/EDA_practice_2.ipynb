```python
plt.figure(figsize=(20,20))
sns.barplot(y = df['Investment Amount (In Lakhs INR) '], x = df['Brand'])
plt.xticks(rotation = 90)
plt.show()
```

```python
df_debt = df[df['Debt (In lakhs INR)'] != 0]
sns.barplot(y = df_debt['Debt (In lakhs INR)'], x = df_debt['Brand'])
plt.xticks(rotation = 90)
plt.show()
```

```python
df_Aman = df[df['Aman'] == 'Y']
df_Anupam = df[df['Anupam'] == 'Y']
df_Namita = df[df['Namita'] == 'Y']
df_Ashneer = df[df['Ashneer'] == 'Y']
df_Peyush = df[df['Peyush'] == 'Y']
df_Vineeta = df[df['Vineeta'] == 'Y']
df_Ghazal = df[df['Ghazal'] == 'Y']
```

```python
sns.scatterplot(x = df_Aman['Equity'],y = df['Investment Amount (In Lakhs INR) '])
plt.xticks(rotation = 90)
plt.title("Aman Gupta investments")
plt.show()
```

```python
sns.scatterplot(x = df_Ghazal['Equity'],y = df['Investment Amount (In Lakhs INR) '])
plt.xticks(rotation = 90)
plt.title("Ghazal Alagh investments")
plt.show()
```

```python
plt.figure(figsize=(12,20))
plt.barh(width = df['Investment Amount (In Lakhs INR) '], y = df['Equity'])
plt.xticks(rotation = 90)
plt.yticks(rotation = 0)
plt.title("Invested Amount in lakhs v/s Equity")
plt.xlabel('Invested Amount')
plt.ylabel('Equity')
plt.show()
```
